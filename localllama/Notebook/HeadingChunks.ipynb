```python
import fitz
```

```python
doc = fitz.open("Reports/tcs.pdf")
```

```python
tcspage=doc[1]
```

```python
blocks = tcspage.get_text("dict")["blocks"]
```

```python
tcspage.get_fonts()
```

```python
chunks =[]
flag = "" #heading or normal
i=0
for b in blocks:  # iterate through the text blocks
   if "lines" in b: 
    for l in b["lines"]:  # iterate through the text lines
        for s in l["spans"]:  # iterate through the text spans
            if ("Bold" in s["font"] or "DemiB" in s["font"]) and len(l["spans"])==1:
            #  if '\n' in s["text"]:
                chunks.append(s["text"])
                if flag == "normal":
                    i=i+1
                    flag = "heading"
                # print(s['text'], s['size'] , s["font"])
            else:
                if flag == "":
                    chunks.append(s["text"])
                chunks[i] = chunks[i] +" " + s["text"]
                flag = "normal"
```

```python
docchunks =[]
flag = "" #heading or normal
i=0
for page_num in range(doc.page_count): 
 page = doc.load_page(page_num)
 blocks = page.get_text('dict')["blocks"]
 for b in blocks:  # iterate through the text blocks
   if "lines" in b: 
    for l in b["lines"]:  # iterate through the text lines
        for s in l["spans"]:  # iterate through the text spans
            if ("Bold" in s["font"] or "DemiB" in s["font"]) and len(l["spans"])==1 and (flag == "" or flag!="heading"):
            #  if '\n' in s["text"]:
                docchunks.append(s["text"])
                if flag == "normal":
                    i=i+1
                    flag = "heading"
                # print(s['text'], s['size'] , s["font"])
            else:
                if flag == "":
                    docchunks.append(s["text"])
                docchunks[i] = docchunks[i] +" " + s["text"]
                flag = "normal"
```

```python
def getTextByBlock(page):
    txt=""
    blks = page.get_text("dict")["blocks"]
    for b in blks:  # iterate through the text blocks
        if "lines" in b: 
            for l in b["lines"]:  # iterate through the text lines
                for s in l["spans"]:  # iterate through the text spans
                    txt+= " "+s['text']
    print(txt)
    
    
```
